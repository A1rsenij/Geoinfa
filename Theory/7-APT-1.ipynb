### Метеорологические искусственные спутники Земли серии NOAA

Эксперименты по использованию искусственных спутников Земли (ИСЗ) в метеорологии начались в США в 1960 г. с запуска 1 апреля «TIROS-1» (Television InfraRed Observation Satellite). К 1965 г. было запущено уже 10 таких ИСЗ, которые позволили получить свыше 650 тыс. изображений подстилающей поверхности Земли.

Программа TIROS была создана, чтобы ответить на вопрос: "Могут ли спутники быть полезными в изучении Земли?". В то время эффективность спутниковых наблюдений еще не была доказана. Поскольку спутники были новой технологией, в рамках программы TIROS также проверялись различные вопросы конструкции космических аппаратов: было испытано метеорологическое оборудование, перспективное с точки зрения применения на будущих операционных метеорологических ИСЗ. Цель состояла в том, чтобы улучшить спутниковые приложения для принятия решений, связанных с Землей. Первоочередной задачей программы была разработка метеорологической спутниковой информационной системы. Прогнозирование погоды считалось наиболее многообещающим применением космических наблюдений - с помощью ИСЗ серии TIROS ученые пытались получить метеорологические данные, которые позволят прояснить механизм образования ураганов, тайфунов, циклонов и т.п.

<center><i>Одно из первых изображений Земли из космоса, переданное с TIROS-1</i></center>

Пуски «TIROS-1...10» (1960-65гг.) позволили получить изображения и изучить характеристики земной поверхности и облачного покрова. 

ITOS (Improved TIROS Operational System) стал продолжением серии метеорологических спутников TIROS и положил начало использованию обозначения "NOAA", так как управляющей организацией программы стало Национальное управление океанических и атмосферных исследований (National Oceanic and Atmospheric Administration - NOAA). 

Конструкция ИСЗ серии «NOAA» допускает размещение в ней дополнительного оборудования, что позволяет устанавливать на борту ретрансляторы и процессоры системы поиска и спасения, радиометр обратного рассеяния солнечного УФ-излучения и сканирующий прибор для измерения баланса излучения Земли.

С помощью снимков, переданных ИСЗ серии «NOAA», метеорологи установили, что структура облачного покрова носит глобальный характер и что в любой момент времени от половины до двух третьих поверхности Земли скрыто под облаками. Информация, полученная с ИСЗ серии «NOAA», позволила лучше понять механизм образования и распределения облаков различных типов; кроме того, эти данные позволили обнаружить многие ураганы, шквалы, торнадо и другие явления задолго до того, как они были обнаружены с помощью обычных метеорологических методов.

### Сканирующие радиометры

В настоящее время в оперативной работе используются изображения, получаемые с российских метеорологических ИСЗ серии «Метеор», экспериментальных ИСЗ «Монитор-Э», природно ресурсных ИСЗ «Ресурс», а также американских метеорологических ИСЗ серии «NOAA» — 17 и 19.

NOAA-N Prime (NOAA-19) оснащен набором из восьми инструментов, которые предоставляют данные для прогнозирования погоды и климата. Как и его предшественники, NOAA-N Prime предоставляет глобальные изображения облаков и особенностей поверхности, а также вертикальные профили температуры и влажности атмосферы для использования в численных моделях прогноза погоды и океана, а также данные о распределении озона в верхних слоях атмосферы и околоземной космической среде — информация, важная для морского, авиационного, энергетического, сельскохозяйственного и других сообществ. 

Нам интересен один из основных инструментов NOAA-N Prime — усовершенствованный радиометр очень высокого разрешения (AVHRR/3). AVHRR/3 представляет собой шестиканальный радиометр, который обнаруживает энергию в видимой и инфракрасной (ИК) частях электромагнитного спектра. Прибор измеряет отраженную солнечную (видимую и ближнюю ИК) энергию и излучаемую тепловую энергию поверхности Земли, моря, облаков и промежуточной атмосферы.

*Комментарий: передача данных осуществляется только по 5 каналам, канал 3A (видимый диапазон) работает в дневное время и 3B (инфракрасный) в ночное.*

Прибор сканирует Землю серией линий. Линии ориентированы перпендикулярно направлению движения сенсорной платформы (т. е. поперек полосы). Каждая линия сканируется от одной стороны датчика к другой с помощью вращающегося зеркала **А**. По мере того, как платформа движется над Землей, последовательные сканирования создают двумерное изображение поверхности Земли. Приходящее отраженное или испускаемое излучение разделяется на несколько спектральных составляющих, которые обнаруживаются независимо. УФ, видимое, ближнее инфракрасное и тепловое излучение рассеиваются по составляющим их длинам волн. Группа внутренних детекторов **В**, каждый из которых чувствителен к определенному диапазону длин волн, обнаруживает и измеряет энергию для каждой спектральной полосы, а затем в виде электрического сигнала они преобразуются в цифровые данные и записываются для последующей компьютерной обработки.

### Передача данных о температуре подстилающей поверхности земли

Представим, что у нас на низкой околоземной орбите летает инфракрасный радиометр, который может измерять излучение с поверхности Земли. Для компактности передаваемых данных сожмём результат измерений до величины одного байта. То есть спутник может сообщить нам число от 0 до 255, в зависимости от интенсивности измеренного излучения. 

Вопрос: каких данных не хватает для вычисления температуры подстилающей поверхности земли?<br>
Ответ: двух контрольных измерений, согласно которым будет производиться калибровка радиометра. 

В качестве контрольных измерений часто приходит на ум Солнце или любая большая однородная поверхность на Земле. Спутник может навестись на такую поверхность или на Солнце и измерить показания.

Допустим мы выбрали в качестве контрольной точки измерения Солнце. Тогда радиометр должен обрабатывать значения температур вплоть до 6000К. Температуры на Земле составляют около 300К, поэтому на информационный байт будет приходиться $(6000 \ –  \ 300) \ / \ 255 ≈ 22,35.$

То есть на один информационный бит приходится примерно 22 градуса. Именно это значение будет отвечать за точность измерения. Очевидно, что эта погрешность слишком велика. Чтобы её уменьшить необходимо брать контрольные точки с температурами близкими к 300К.

На спутниках серии NOAA для этого используется бортовое абсолютно черное тело (АЧТ), представленное в виде четырёх платиновых терморезисторов. Их температура в рабочем состоянии составляет примерно 282К.

Вторая контрольная точка – это температура реликтового излучения. По справочной литературе она составляет около 3К.

Так как на Земле существует температура выше чем 282К, то для оценки точности измерения необходимо взять максимально необходимую температуру на Земле. Допустим, что нас не интересуют температуры выше 330К, тогда на информационный байт радиометра будет приходиться $(330 \ – \ 3) \ / \ 255 ≈ 1,28$. Такая точность нас вполне устроит. 

Обозначим измеренное радиометром значение яркостной температуры Земли в условных единицах. Чтобы получить термодинамическую температуру, необходимо знать все значения контрольных точек. А именно: яркостную температуру космоса в условных единицах, термодинамическую температуру космоса, яркостную температуру АЧТ в условных единицах, термодинамическую температуру АЧТ. Среди этих величин только термодинамическая температура космоса неизменна, остальные же величины должны быть переданы со спутника для калибровки изображения. 

В итоге имеем, что от нашего радиометра на орбите для его калибровки, необходимо передавать данные об измерении температуры космоса радиометром, данные об измерении излучения и температуры АЧТ.

### Система автоматической передачи изображений (automatic picture transmission - APT)

Система автоматической передачи изображений (APT) представляет собой аналоговую систему передачи изображений, разработанную для использования на метеорологических спутниках. Была внедрена в 60-х годах и уже более пяти десятилетий обеспечивает получение метеоснимков на относительно простых приёмных станциях по всему миру. Приём локальных данных на любой станции возможен как минимум дважды в сутки с каждого спутника, при его проходе над станцией. В APT спутник передаёт непрерывную ленту изображения той местности, над которой пролетает. В изображении всего два спектральных канала, поэтому изображение не является цветным в обычном понимании. Окрашивание снимка и наложение контуров крупных ориентиров иногда производятся при постобработке для упрощения чтения снимка.

<center><img src="https://www.nasa.gov/images/content/315000main_First_Image600.jpg" alt="NOAA-19 перед запуском" style="width: 400px;"/></center>
<center><i>Первое изображение с NOAA-19</i></center>

Передача снимка осуществляется построчно. В каждую строку горизонтальной развёртки входят данные двух каналов изображения (обычно обозначаются Video A и Video B), данные телеметрии, специальные последовательности для синхронизации. Длина каждой строки изображения — 2080 пикселей. Каждый канал изображения занимает по 909 пикселей, остаток используется для телеметрии и синхроданных. В секунду передаётся 2 строки.

*Комментарий: спутники семейства NOAA, используют APT для передачи изображений разрешением 4 км/пиксель и разрядностью 8 бит, получаемые ресемлпированием изображений с AVHRR. Перед передачей изображения геометрически корректируются, то есть они свободны от искажений, вызванных кривизной Земли. NOAA может настроить спутник для передачи в APT любых двух спектральных каналов радиометра. Чаще всего, в одном канале непрерывно передаётся изображение с длинноволнового ИК-канала, а в другом канале на освещённой стороне Земли передаётся изображение в ближнем ИК диапазоне, а на не освещённой — в среднем ИК-диапазоне.*
